```python
text="Hoy es un dia mas loco que he tenido, quiero una manzana Cuando estamos enamorados, lo único que queremos es declarar nuestro amor a los cuatro vientos. Expresar este sentimiento requiere mucha valentía.Hemos seleccionado los mejores textos, mensajes y declaraciones de amor para que le envíes a tu novio por las redes sociales, se los escribas en una linda nota o se los digas en voz alta La mejor parte de mi día es tu sonrisa.Estar contigo fue la decisión más fácil que tomé en la vida y me siento la persona más afortunada del mundo todos los días.Me encanta tu honestidad, tu pasión, tu resiliencia, tu valentía, tu compañerismo. Eres mi alma gemela, mi mejor amigo y mi confidente.Me gustaría que pudieras verte todos los días a través de mis ojos, porque te veo como una persona perfecta. Mi vida fue en blanco y negro. Llegaste y la coloreaste con tu sonrisa y tu luz."
doc = nlp(text)
for token in doc:
    print(token.text, token.pos_, token.dep_)
```
